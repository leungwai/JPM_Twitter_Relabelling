```python
import pandas as pd
```

```python
lw_val = pd.read_csv('Leung Wai - VAL_SINGLE_twitter_label - VAL_SINGLE_twitter_label.tsv.tsv', sep='\t')
lw_val
```

```python
lw_val_FINAL = lw_val.copy()


for index, row in lw_val_FINAL.iterrows(): 
    proposed_ner_label = row['PROPOSED NER Label']

    if proposed_ner_label == "['B-PER']":
        lw_val_FINAL.loc[index, 'PROPOSED NER Label'] = "B-PER"
    elif proposed_ner_label == "['B-ORG']":
        lw_val_FINAL.loc[index, 'PROPOSED NER Label'] = "B-ORG"
    elif proposed_ner_label == "['B-LOC']":
        lw_val_FINAL.loc[index, 'PROPOSED NER Label'] = "B-LOC"
    elif proposed_ner_label == "['B-MISC']":
        lw_val_FINAL.loc[index, 'PROPOSED NER Label'] = "B-MISC"
    elif proposed_ner_label == "['O']":
        lw_val_FINAL.loc[index, 'PROPOSED NER Label'] = "O"

    actual_ner_label = row['LW NER Label']

    if actual_ner_label == "[B-PER]":
        lw_val_FINAL.loc[index, 'LW NER Label'] = 'B-PER'
    elif actual_ner_label == "[B-ORG]":
        lw_val_FINAL.loc[index, 'LW NER Label'] = 'B-ORG'
    elif actual_ner_label == "[B-LOC]":
        lw_val_FINAL.loc[index, 'LW NER Label'] = 'B-LOC'
    elif actual_ner_label == "[B-MISC]":
        lw_val_FINAL.loc[index, 'LW NER Label'] = 'B-MISC'
    elif actual_ner_label == "[O]":
        lw_val_FINAL.loc[index, 'LW NER Label'] = 'O'
    
lw_val_FINAL
```

```python
# Removing the date

spacy_date_labelling = "'DATE'|'TIME'" 
filter = lw_val_FINAL['Spacy NER Label'].str.contains(spacy_date_labelling)

lw_val_FINAL = lw_val_FINAL[~filter]

spacy_date_labelling_two = "am|pm|jan|feb|mar|apr|may|june|july|aug|sept|oct|nov|dec|january|february|march|april|may|june|july|august|september|october|november|december|mon|tues|wed|thur|thurs|fri|sat|sun|monday|tuesday|wednesday|thursday|friday|saturday|sunday"
filter_two = lw_val_FINAL['Word'].str.contains(spacy_date_labelling_two, case=False)

lw_val_FINAL = lw_val_FINAL[~filter_two]

lw_val_FINAL
```
