```python
import pandas as pd
```

```python
lw_test = pd.read_csv('Leung Wai - TEST_SINGLE_twitter_label - TEST_SINGLE_twitter_label.tsv.tsv', sep='\t')
lw_test
```

```python
lw_test_FINAL = lw_test.copy()


for index, row in lw_test_FINAL.iterrows(): 
    proposed_ner_label = row['PROPOSED NER Label']

    if proposed_ner_label == "['B-PER']":
        lw_test_FINAL.loc[index, 'PROPOSED NER Label'] = "B-PER"
    elif proposed_ner_label == "['B-ORG']":
        lw_test_FINAL.loc[index, 'PROPOSED NER Label'] = "B-ORG"
    elif proposed_ner_label == "['B-LOC']":
        lw_test_FINAL.loc[index, 'PROPOSED NER Label'] = "B-LOC"
    elif proposed_ner_label == "['B-MISC']":
        lw_test_FINAL.loc[index, 'PROPOSED NER Label'] = "B-MISC"
    elif proposed_ner_label == "['O']":
        lw_test_FINAL.loc[index, 'PROPOSED NER Label'] = "O"

    actual_ner_label = row['LW NER LABEL']

    if actual_ner_label == "[B-PER]":
        lw_test_FINAL.loc[index, 'LW NER LABEL'] = 'B-PER'
    elif actual_ner_label == "[B-ORG]":
        lw_test_FINAL.loc[index, 'LW NER LABEL'] = 'B-ORG'
    elif actual_ner_label == "[B-LOC]":
        lw_test_FINAL.loc[index, 'LW NER LABEL'] = 'B-LOC'
    elif actual_ner_label == "[B-MISC]":
        lw_test_FINAL.loc[index, 'LW NER LABEL'] = 'B-MISC'
    elif actual_ner_label == "[O]":
        lw_test_FINAL.loc[index, 'LW NER LABEL'] = 'O'
    
lw_test_FINAL
```

```python
# Removing the date

spacy_date_labelling = "'DATE'|'TIME'" 
filter = lw_test_FINAL['Spacy NER Label'].str.contains(spacy_date_labelling)

lw_test_FINAL = lw_test_FINAL[~filter]

spacy_date_labelling_two = "am|pm|jan|feb|mar|apr|may|june|july|aug|sept|oct|nov|dec|january|february|march|april|may|june|july|august|september|october|november|december|mon|tues|wed|thur|thurs|fri|sat|sun|monday|tuesday|wednesday|thursday|friday|saturday|sunday"
filter_two = lw_test_FINAL['Word'].str.contains(spacy_date_labelling_two, case=False)

lw_test_FINAL = lw_test_FINAL[~filter_two]

lw_test_FINAL
```
