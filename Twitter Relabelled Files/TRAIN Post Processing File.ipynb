```python
import pandas as pd
```

```python
lw_train = pd.read_csv('Leung Wai - TRAIN_SINGLE_twitter_label - TRAIN_SINGLE_twitter_label.tsv.tsv', sep='\t')
lw_train
```

```python
lw_train_FINAL = lw_train.copy()


for index, row in lw_train_FINAL.iterrows(): 
    proposed_ner_label = row['PROPOSED NER Label']

    if proposed_ner_label == "['B-PER']":
        lw_train_FINAL.loc[index, 'PROPOSED NER Label'] = "B-PER"
    elif proposed_ner_label == "['B-ORG']":
        lw_train_FINAL.loc[index, 'PROPOSED NER Label'] = "B-ORG"
    elif proposed_ner_label == "['B-LOC']":
        lw_train_FINAL.loc[index, 'PROPOSED NER Label'] = "B-LOC"
    elif proposed_ner_label == "['B-MISC']":
        lw_train_FINAL.loc[index, 'PROPOSED NER Label'] = "B-MISC"
    elif proposed_ner_label == "['O']":
        lw_train_FINAL.loc[index, 'PROPOSED NER Label'] = "O"

    actual_ner_label = row['LW NER Label']

    if actual_ner_label == "[B-PER]":
        lw_train_FINAL.loc[index, 'LW NER Label'] = 'B-PER'
    elif actual_ner_label == "[B-ORG]":
        lw_train_FINAL.loc[index, 'LW NER Label'] = 'B-ORG'
    elif actual_ner_label == "[B-LOC]":
        lw_train_FINAL.loc[index, 'LW NER Label'] = 'B-LOC'
    elif actual_ner_label == "[B-MISC]":
        lw_train_FINAL.loc[index, 'LW NER Label'] = 'B-MISC'
    elif actual_ner_label == "[O]":
        lw_train_FINAL.loc[index, 'LW NER Label'] = 'O'
    
lw_train_FINAL
```

```python
# Removing the date

spacy_date_labelling = "'DATE'|'TIME'" 
filter = lw_train_FINAL['Spacy NER Label'].str.contains(spacy_date_labelling)

lw_train_FINAL = lw_train_FINAL[~filter]

spacy_date_labelling_two = "am|pm|jan|feb|mar|apr|may|june|july|aug|sept|oct|nov|dec|january|february|march|april|may|june|july|august|september|october|november|december|mon|tues|wed|thur|thurs|fri|sat|sun|monday|tuesday|wednesday|thursday|friday|saturday|sunday"
filter_two = lw_train_FINAL['Word'].str.contains(spacy_date_labelling_two, case=False)

lw_train_FINAL = lw_train_FINAL[~filter_two]

lw_train_FINAL
```

```python
alec_train = pd.read_csv('Alec - TRAIN_SINGLE_twitter_label - TRAIN_SINGLE_twitter_label.tsv.tsv', sep='\t')
alec_train
```

```python
alec_train_FINAL = alec_train.copy()


for index, row in alec_train_FINAL.iterrows(): 
    proposed_ner_label = row['PROPOSED NER Label']

    if proposed_ner_label == "['B-PER']":
        alec_train_FINAL.loc[index, 'PROPOSED NER Label'] = "B-PER"
    elif proposed_ner_label == "['B-ORG']":
        alec_train_FINAL.loc[index, 'PROPOSED NER Label'] = "B-ORG"
    elif proposed_ner_label == "['B-LOC']":
        alec_train_FINAL.loc[index, 'PROPOSED NER Label'] = "B-LOC"
    elif proposed_ner_label == "['B-MISC']":
        alec_train_FINAL.loc[index, 'PROPOSED NER Label'] = "B-MISC"
    elif proposed_ner_label == "['O']":
        alec_train_FINAL.loc[index, 'PROPOSED NER Label'] = "O"

    actual_ner_label = row['Alec NER Label']

    if actual_ner_label == "[B-PER]":
        alec_train_FINAL.loc[index, 'Alec NER Label'] = 'B-PER'
    elif actual_ner_label == "[B-ORG]":
        alec_train_FINAL.loc[index, 'Alec NER Label'] = 'B-ORG'
    elif actual_ner_label == "[B-LOC]":
        alec_train_FINAL.loc[index, 'Alec NER Label'] = 'B-LOC'
    elif actual_ner_label == "[B-MISC]":
        alec_train_FINAL.loc[index, 'Alec NER Label'] = 'B-MISC'
    elif actual_ner_label == "[O]":
        alec_train_FINAL.loc[index, 'Alec NER Label'] = 'O'
    
alec_train_FINAL
```

```python
# Removing the date

spacy_date_labelling = "'DATE'|'TIME'" 
filter = alec_train_FINAL['Spacy NER Label'].str.contains(spacy_date_labelling)

alec_train_FINAL = alec_train_FINAL[~filter]

spacy_date_labelling_two = "am|pm|jan|feb|mar|apr|may|june|july|aug|sept|oct|nov|dec|january|february|march|april|may|june|july|august|september|october|november|december|mon|tues|wed|thur|thurs|fri|sat|sun|monday|tuesday|wednesday|thursday|friday|saturday|sunday"
filter_two = alec_train_FINAL['Word'].str.contains(spacy_date_labelling_two, case=False)

alec_train_FINAL = alec_train_FINAL[~filter_two]

alec_train_FINAL
```

```python
combined_train_final = lw_train_FINAL.copy()
combined_train_final
```

```python
#adding Alec's row to combined_row_final

alec_train_row = alec_train_FINAL['Alec NER Label'].tolist()
combined_train_final.insert(12, 'Alec NER Label', alec_train_row)

combined_train_final
```

```python
combined_train_agree = combined_train_final.drop(combined_train_final[combined_train_final['LW NER Label'] != combined_train_final['Alec NER Label']].index)
combined_train_agree
```

```python
list_hierarchy = ['B-PER', 'B-ORG', 'B-LOC', 'B-MISC', 'O']
combined_train_agree['PROPOSED NER Label'] = pd.Categorical(combined_train_agree['PROPOSED NER Label'], ordered=True, categories=list_hierarchy)
combined_train_agree_filtered = combined_train_agree.sort_values('PROPOSED NER Label')

combined_train_agree_filtered
```

```python
combined_train_disagree = combined_train_final.drop(combined_train_final[combined_train_final['LW NER Label'] == combined_train_final['Alec NER Label']].index)
combined_train_disagree
```

```python
list_hierarchy = ['B-PER', 'B-ORG', 'B-LOC', 'B-MISC', 'O']
combined_train_disagree['PROPOSED NER Label'] = pd.Categorical(combined_train_disagree['PROPOSED NER Label'], ordered=True, categories=list_hierarchy)
combined_train_disagree_filtered = combined_train_disagree.sort_values('PROPOSED NER Label')

combined_train_disagree_filtered
```
