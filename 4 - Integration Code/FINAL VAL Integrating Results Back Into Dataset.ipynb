```python
# WRITING BACK TO ORIGINAL FILE
filename = '../Datasets/NER/Tweebank/' + 'dev.bio'

how_much_changed = 0
replacement = ''
with open(filename, 'r') as file:
    original_data = file.readlines()

with open(filename, 'r') as file:
    sentence_count = 0
    word_in_sentence_count = 0
    
    for i, line in enumerate(file):
        line = line[:-1]

        if len(line):
            check_df = original_dataset_merged[original_dataset_merged['Sentence Index'] == sentence_count]
            
            if len(check_df) != 0: 
                check_df_row = check_df[check_df['Start Word Index'] == word_in_sentence_count]
                
                if len(check_df_row) != 0:
                    check_df_row_value = check_df_row['FINAL NER Label']
                    check_df_row_value_empty = check_df_row_value.any()

                    if check_df_row_value_empty != False:
                        token, _, _, label = line.split('\t')
                        
                        print("-" * 50)
                        print("ORIGINAL LINE")
                        print(line)
                        print("WORD IN TOKEN")
                        print(check_df_row.iloc[0]['Word'])
                        line_to_replace = [token, '-', '-', check_df_row.iloc[0]['FINAL NER Label']]
                        line_to_replace_string = '\t'.join(line_to_replace)
                        print("Line to replace")
                        print(line_to_replace_string) 

                        original_data[i] = line_to_replace_string + '\n'
                        how_much_changed += 1
        
            word_in_sentence_count += 1
        else:
            sentence_count += 1
            word_in_sentence_count = 0



to_write_file = open('FINAL_dev.bio', 'w')
to_write_file.writelines(original_data)
to_write_file.close()

print("HOW MUCH CHANGED:")
print(how_much_changed)
```

```python
## SANITY CHECK FINAL DATASET
filename = '../Datasets/NER/Tweebank/' + 'dev.bio'

original_data_check = []
new_data_check = []

total_changed_additions = 0
total_new_additions = 0

total_old_entities = 0
total_new_entities = 0


with open(filename, 'r') as file:
    original_data_check = file.readlines()

with open('FINAL_dev.bio', 'r') as file_two:
    new_data_check = file_two.readlines()

print(len(original_data_check))
print(len(new_data_check))

for i in range(len(original_data_check)):
    original_string = original_data_check[i]
    new_string = new_data_check[i]
    
    if original_string != '\n':
        old_token, _, _, old_label = original_string.split('\t')
        new_token, _, _, new_label = new_string.split('\t')
        
        if old_label != "O\n":
            total_old_entities += 1
        if new_label != "O\n":
            total_new_entities += 1

        if old_label != new_label:
            print("-" * 50)
            print("CHANGED ADDITION")
            print("OLD STRING")
            print(original_string)
            print("NEW STRING")
            print(new_string)
            print("OLD LABEL:", old_label)
            print("NEW LABEL:", new_label)
            total_changed_additions += 1
        if (old_label == "O\n") and (new_label != "O\n"):
            print("ALSO NEW ADDITION AS WELL")
            total_new_additions += 1
            
print("TOTAL OLD ENTITIES", total_old_entities)
print("TOTAL NEW ENTITIES", total_new_entities)
print("TOTAL CHANGED ADDITIONS", total_changed_additions)            
print("TOTAL NEW ADDITIONS", total_new_additions)

```
