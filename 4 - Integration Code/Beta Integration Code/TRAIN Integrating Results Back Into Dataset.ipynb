```python
import pandas as pd
import numpy as np
```

```python
original_dataset = pd.read_csv('logs/combined_train_labels_FINAL_NUMBERS.tsv', sep='\t')
original_dataset
```

```python
original_dataset.index.name='Orig Index'
original_dataset
```

```python
train_agree = pd.read_csv('../../JPM_Twitter_Relabelling/Twitter Relabelled Files/Combined Train Files/combined_train_agree.tsv', sep='\t')
train_agree
```

```python
train_agree = train_agree.rename(columns={'Unnamed: 0': 'Orig Index'})
train_agree
```

```python
train_disagree = pd.read_csv('../../JPM_Twitter_Relabelling/Twitter Relabelled Files/Combined Train Files/combined_train_disagree_filtered.tsv', sep='\t')
train_disagree
```

```python
train_disagree = train_disagree.rename(columns={'Unnamed: 0': 'Orig Index'})
train_disagree
```

```python
train_relabelled_frames = [train_agree, train_disagree]
train_relabelled = pd.concat(train_relabelled_frames)
train_relabelled
```

```python
train_relabelled = train_relabelled[train_relabelled['Word'].str.contains('@user') == False]
train_relabelled
```

```python
#SANITY CHECK
train_relabelled_proposed_correct = train_relabelled[train_relabelled['PROPOSED NER Label'] != train_relabelled['LW NER Label']]
train_relabelled_proposed_correct
```

```python
original_dataset_merged = pd.merge(original_dataset, train_relabelled[['Orig Index', 'LW NER Label']], on='Orig Index', how='left')
original_dataset_merged
```

```python
original_dataset_merged = original_dataset_merged.rename(columns={'LW NER Label' : 'FINAL NER Label'})
original_dataset_merged
```

```python
original_dataset_merged = original_dataset_merged[original_dataset_merged['Word'].str.contains('@user') == False]
original_dataset_merged
```

```python
original_dataset_merged.to_csv('logs/original_dataset_merged_with_results.tsv', sep='\t')
```

```python
# WRITING BACK TO ORIGINAL FILE
filename = '../Datasets/NER/Tweebank/' + 'train copy.bio'

how_much_changed = 0
replacement = ''
with open(filename, 'r') as file:
    original_data = file.readlines()

with open(filename, 'r') as file:
    sentence_count = 0
    word_in_sentence_count = 0
    
    for i, line in enumerate(file):
        line = line[:-1]

        if len(line):
            check_df = original_dataset_merged[original_dataset_merged['Sentence Index'] == sentence_count]
            
            if len(check_df) != 0: 
                check_df_row = check_df[check_df['Start Word Index'] == word_in_sentence_count]
                
                if len(check_df_row) != 0:
                    check_df_row_value = check_df_row['FINAL NER Label']
                    check_df_row_value_empty = check_df_row_value.any()

                    if check_df_row_value_empty != False:
                        token, _, _, label = line.split('\t')
                        
                        print("-" * 50)
                        print("ORIGINAL LINE")
                        print(line)
                        print("WORD IN TOKEN")
                        print(check_df_row.iloc[0]['Word'])
                        if (token.lower() == (check_df_row.iloc[0]['Word']).lower()):
                            line_to_replace = [token, '-', '-', check_df_row.iloc[0]['FINAL NER Label']]
                            line_to_replace_string = '\t'.join(line_to_replace)
                            print("Line to replace")
                            print(line_to_replace_string) 
                            original_data[i] = line_to_replace_string + '\n'
                            how_much_changed += 1 
        
            word_in_sentence_count += 1
        else:
            sentence_count += 1
            word_in_sentence_count = 0



to_write_file = open('train_test.bio', 'w')
to_write_file.writelines(original_data)
to_write_file.close()

print("HOW MUCH CHANGED:")
print(how_much_changed)
```
